# ***4. Unit Roots***

```python
from statsmodels.tsa.stattools import adfuller
import ipeadatapy
```

First we need to import some datas for to use like exemple. 

```python
Sudeste = ipeadatapy.timeseries("ELETRO12_CEESE12")
Sudeste["VALUE (GWh)"].plot()
```

#### **4.1 Dickey-Fuller Test**

```python
adfuller(Sudeste["VALUE (GWh)"], maxlag= 1, store = True, regresults = True)
```

```python
adfuller(Sudeste["VALUE (GWh)"].diff()[1:], maxlag= 1, store = True, regresults = True)
```

```python
adfuller(Sudeste["VALUE (GWh)"], maxlag= 4, store = True, regresults = True)
```
